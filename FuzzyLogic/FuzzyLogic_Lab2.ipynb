```python
import numpy as np
import skfuzzy as fuzz

# Tập U gồm 5 ứng cử viên vào chức vụ quản đốc phân xưởng
U = ('U1', 'U2', 'U3', 'U4', 'U5')

# Định nghĩa các biến đàu vào
degree_of_skill = np.arange(0, 1.1, 0.1) # Tiêu chuẩn A- trình độ NV
years_of_experience = np.arange(0, 1.1, 0.1) # Tiêu chuẩn B- kinh nghiệm

# Định nghĩa hàm xấp xỉ Fuzz cho mức độ phù hợp của từng ứng viên với tiêu chuẩn A
membership_degree_A = {
    'U1': fuzz.trimf(degree_of_skill, (0.4, 0.4, 1)),
    'U2': fuzz.trimf(degree_of_skill, (1, 1, 1)),
    'U3': fuzz.trimf(degree_of_skill, (0.8, 1, 1)),
    'U4': fuzz.trimf(degree_of_skill, (0.2, 0.2, 0.6)),
    'U5': fuzz.trimf(degree_of_skill, (0.4, 0.4, 0.8)),
}

# Định nghĩa hàm xấp xỉ Fuzzy cho mức độ p/hợp của từng ứng viên với tiêu chuẩn B
membership_degree_B = {
    'U1': fuzz.trimf(years_of_experience, (0.4, 0.4, 0.8)),
    'U2': fuzz.trimf(years_of_experience, (0.6, 1, 1)),
    'U3': fuzz.trimf(years_of_experience, (1, 1, 1)),
    'U4': fuzz.trimf(years_of_experience, (0, 0, 0.2)),
    'U5': fuzz.trimf(years_of_experience, (0.8, 1, 1)),
}

# Tìm tập con mờ của U những ứng viên thỏa ít nhất 1/2 tiêu chuẩn A/B
membership_at_least_one = {}
membership_at_both = {}
membership_at_degree_B_only = {}
for u in U:
  membership_at_least_one[u] = np.fmax(membership_degree_A[u], membership_degree_B[u])
for u in U:
  membership_at_both[u] = np.fmin(membership_degree_A[u], membership_degree_B[u])
for u in U:
  membership_at_degree_B_only[u] = np.max(membership_degree_A[u])

# In KQ
print('=> Ứng viên thỏa ít nhất 1/2 tiêu chuẩn A/B')
for u in U:
  print(f'Ứng viên {u}:', membership_at_least_one[u])

print('=> Ứng viên thỏa cả 2 tiêu chuẩn A/B')
for u in U:
  print(f'Ứng viên {u}:', membership_at_both[u])

print('=> Ứng viên kg thỏa tiêu chuẩn A')
for u in U:
  print(f'Ứng viên {u}:', membership_at_degree_B_only[u])

print('=> Nhát cắt mức Alpha=0.6 của tập mờ là:')
for u in U:
  print(fuzz.lambda_cut(ms=membership_at_both[u], lcut=0.6))
print('=> Nhát cắt mức Alpha=0.8 của tập mờ là:')
for u in U:
  print(fuzz.lambda_cut(ms=membership_at_both[u], lcut=0.8))


```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Định nghĩa các biến đầu vào
nhiet_do = ctrl.Antecedent(np.arange(0, 11, 1), 'nhiet_do')

# Định nghĩa các biến đầu ra
muc_do = ctrl.Consequent(np.arange(0, 11, 1), 'muc_do')

# Xác định các hàm xấp xỉ Fuzzy
nhiet_do['low'] = fuzz.trimf(nhiet_do.universe, (0, 0, 10))
nhiet_do['medium'] = fuzz.trimf(nhiet_do.universe, (0, 5, 10))
nhiet_do['high'] = fuzz.trimf(nhiet_do.universe, (5, 10, 10))
muc_do['low'] = fuzz.trimf(muc_do.universe, (0, 0, 10))
muc_do['medium'] = fuzz.trimf(muc_do.universe, (0, 5, 10))
muc_do['high'] = fuzz.trimf(muc_do.universe, (5, 10, 10))

# Xác định quy tắc Fuzzy
rule1 = ctrl.Rule(nhiet_do['low'], muc_do['low'])
rule2 = ctrl.Rule(nhiet_do['medium'], muc_do['medium'])
rule3 = ctrl.Rule(nhiet_do['high'], muc_do['high'])

# Xây dựng hệ thống Fuzzy Logic
service_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
service = ctrl.ControlSystemSimulation(service_ctrl)

# Đưa ra các giá trị đầu vào
service.input['nhiet_do'] = 100

# Đánh giá chất lượng dịch vụ
service.compute()

# In KQ
print(service.output['muc_do'])
muc_do.view(sim=service)
```
