```python
x = np.arange(1, 5)
y = x**3
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
ax[0, 1].plot([1, 2, 3, 4], [1, 4, 9, 16], 'go')
ax[0, 1].set_title('Squares')
ax[1, 0].plot(x, y, 'r^')
ax[1, 0].set_title('Cubes')
plt.show()
```

```python
divisions = ['div-a', 'div-b', 'div-c', 'div-d', 'div-e']
division_avg_masks = [70, 82, 73, 65, 68]

plt.bar(divisions, division_avg_masks, color='g')
plt.title('Bar Graph')
plt.xlabel('Divisions')
plt.ylabel('Average Masks')
plt.show()
```

```python
divisions = ['div-a', 'div-b', 'div-c', 'div-d', 'div-e']
division_avg_masks = [70, 82, 73, 65, 68]

plt.barh(divisions, division_avg_masks, color='g')
plt.title('Bar Graph')
plt.xlabel('Divisions')
plt.ylabel('Average Masks')
plt.show()
```

```python
divisions = ['div-a', 'div-b', 'div-c', 'div-d', 'div-e']
division_avg_masks = [70, 82, 73, 65, 68]
variance = [5, 8, 7, 6, 4]

plt.bar(divisions, division_avg_masks, xerr=variance, color='g')
plt.title('Bar Graph')
plt.xlabel('Divisions')
plt.ylabel('Average Masks')
plt.show()
```

```python
divisions = ['div-a', 'div-b', 'div-c', 'div-d', 'div-e']
division_avg_masks = [70, 82, 73, 65, 68]
boys_avg_masks = [68, 67, 77, 61, 70]

idx = np.arange(len(divisions))
width = .30

plt.bar(idx, division_avg_masks, width, color='g', label='Division Masks')
plt.bar(idx + width, boys_avg_masks, width, color='r', label='Boys Masks')


plt.xlabel('Divisions')
plt.ylabel('Masks')
plt.xticks(idx+width/2, divisions)

plt.legend(loc='best')
plt.show()
```

```python
divisions = ['div-a', 'div-b', 'div-c', 'div-d', 'div-e']
girls_avg_masks = [72, 97, 69, 69, 66]
boys_avg_masks = [68, 67, 77, 61, 70]

idx = np.arange(len(divisions))
width = .30

plt.bar(idx, girls_avg_masks, width, color='r', label='Girl Masks', bottom=boys_avg_masks)
plt.bar(idx, boys_avg_masks, width, color='b', label='Boys Masks')

plt.title('Vertically Stacked Bar Graphs')
plt.xlabel('Divisions')
plt.ylabel('Masks')
plt.xticks(idx, divisions)

plt.legend(loc='best')
plt.show()
```

```python
firms = ['firm-a', 'firm-b', 'firm-c', 'firm-d', 'firm-e']
market_share = [20, 25, 15, 10, 20]
Explode = [0, .1, 0, 0, 0]

plt.pie(market_share, explode=Explode, labels=firms, shadow=True, startangle=45)
plt.axis('equal')
plt.legend(title='List of Firms')
plt.show()
```

```python
x = np.random.randn(1000)

plt.title('Histogram')
plt.xlabel('Random Data')
plt.ylabel('Frequency')
plt.hist(x, bins=10, color='g', edgecolor='black')
plt.show()
```

```python
height = np.array([167, 170, 149, 165, 155, 180, 166, 146, 159, 185, 145, 168, 172, 181, 169])
weight = np.array([86, 74, 66, 78, 68, 79, 90, 73, 70, 88, 66, 85, 67, 84, 77])

plt.xlim(140, 200)
plt.ylim(60, 100)
plt.scatter(height, weight, color='g')
plt.title('Scatter Plot')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(height, weight)
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(height, weight)
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.show()
```
