```python
import glob2
import numpy as np
import os


!rm -f train.txt valid.txt

all_files = []

for ext in ["*.png", "*.jpg", "*.jpeg"]:
  images = glob2.glob(os.path.join("/content/drive/MyDrive/AI/DL/yolo/yolo_v3/darknet/traindata/img", ext))
  all_files += images

rand_idx = np.random.randint(0, len(all_files), 20)

with open("train.txt", "w") as f:
  for idx in np.arange(len(all_files)):
    f.write(all_files[idx] + "\n")

with open("valid.txt", "w") as f:
  for idx in np.arange(len(all_files)):
    if idx in rand_idx:
      f.write(all_files[idx] + "\n")
```
